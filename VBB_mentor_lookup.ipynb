```python
import pandas as pd
```

```python
export = pd.read_csv("export.csv") #reading the export csv file
```

```python
export.info() #listing all the columns in the file
```

```python
day_week = {0:'00_Mon', 1:'01_Tues', 2:'02_Wed', 3:'03_Thur', 4:'04_Fri', 5:'05_Sat', 6:'06_Sun'}
#Creating mapping for days after MSM calculation
```

```python
export["Program"] = export["COMPUTER_EMAIL"].str[0:4]
#Creating a column called Program using only first 4 characters from computer_email column
```

```python
export.head()
```

```python
#creating columns Day and Time_EST from MSM using calculations from the 30-day attendance excel file
export = export[pd.to_numeric(export['MSM'], errors='coerce').notnull()]
tmp = export['MSM'].astype(int)/1440
export['Time_EST'] = (export['MSM'].astype(int) - 1440*tmp.round())/60
export['Time_EST'] = export['Time_EST'].mod(24).astype(int)

export['Day'] = tmp.round().astype(int).mod(7).replace(day_week)
```

```python
export.head()
```

```python
mentor_lookup = export[["Program","Day","Time_EST"]] 
#Creating a new table by using certain columns from export
```

```python
mentor_lookup.head()
```

```python
mentor_lookup["Mentor_First"] = export["FIRST_NAME"]
mentor_lookup["Mentor_Last"] = export["LAST_NAME"]
mentor_lookup["Mentor_Email"] = export["PERSONAL_EMAIL"]
mentor_lookup["Affiliation"] = export["AFFILIATION"]
#Adding new columns to mentor_lookup from export 
```

```python
mentor_lookup.head()
```
