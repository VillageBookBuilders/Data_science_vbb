```python
#Functions
def average_duration(date:str,participant:str=None):
    if participant is None:
        return (data_meetlogs[(data_meetlogs['Date'] == date)].groupby(['Identify_type']).mean()/60)['Duration']
    else:
        return (data_meetlogs[(data_meetlogs['Date'] == date) & (data_meetlogs['Participant Name'].str.contains(participant))].groupby(['Identify_type']).mean()/60)['Duration']

def get_percentages(array: list):
    a = np.array(array)/sum(array)
    print(f'Values: {list(a)}\n')
    print(f'Percentage: {[str(np.round(x,2))+" %" for x in 100*a]}')
```
