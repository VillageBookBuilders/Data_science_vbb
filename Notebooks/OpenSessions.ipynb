```python
import pandas as pd
from google.colab import drive
from datetime import date, timedelta, datetime
import numpy as np

drive.mount('/content/drive')

# Read CSV file
data_sessions = pd.read_csv('/content/drive/MyDrive/opensessions_cleaned.csv')
```

```python
data_sessions.loc[data_sessions['NAME']=='Adeiso, Ghana']
```

```python
day_week = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}

def find_days_between(start_date, end_date):
    total_days = (end_date-start_date).days
    all_days = [start_date + timedelta(days=x) for x in range(0,total_days,7)]
    return all_days
```

```python
column_names = ["Date", "Day", "Time_EST", "Program", "Paired"]
data = pd.DataFrame(columns = column_names)

data_sessions = data_sessions[pd.to_numeric(data_sessions['MSM'], errors='coerce').notnull()]
data_sessions = data_sessions[~pd.to_numeric(data_sessions['COMPUTER_EMAIL'], errors='coerce').notnull()]
data['Program'] = data_sessions['COMPUTER_EMAIL'].astype(str).str[0:4]
# Drop 'http' from program
data = data[data['Program'].str.contains("http") == False]

tmp = data_sessions['MSM'].astype(int)/1440
data['Time_EST'] = (data_sessions['MSM'].astype(int) - 1440*tmp.round())/60
data['Time_EST'] = data['Time_EST'].astype(int).mod(24).map('{:,.0f}:00'.format)

data['Day'] = tmp.apply(np.floor).mod(7).replace(day_week)

data_sessions['START_DATE'] = pd.to_datetime(data_sessions['START_DATE'], errors='coerce')
data_sessions['END_DATE'] = pd.to_datetime(data_sessions['END_DATE'], errors='coerce')
data['Date'] = data_sessions[['START_DATE','END_DATE']].apply(lambda x: find_days_between(date(int(x[0].year),int(x[0].month),int(x[0].day)),date(int(x[1].year),int(x[1].month),int(x[1].day))) if(pd.notnull(x[0])) else x[0], axis=1)

data["Paired"] = data['Date'].notna()
#data["Paired"] = data_sessions['MENTOR_ID'].notna()
```
