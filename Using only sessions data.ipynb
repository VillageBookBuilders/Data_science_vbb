```python
import pandas as pd
```

```python
sessions_data = pd.read_csv("Sessions Data.csv")
```

```python
sessions_data.head()
```

```python
required_columns = sessions_data[["Program","Day","Time EST","Mentor First","Mentor Last","Mentor Email","Affliation","Monthly Hours"]]
```

```python
required_columns.head()
```

```python
df1 = required_columns.groupby(['Program','Day','Time EST','Mentor First','Mentor Last','Mentor Email']).sum()["Monthly Hours"]
```

```python
df1.head()
```

```python
df1.to_csv("only sessions data.csv")
```
